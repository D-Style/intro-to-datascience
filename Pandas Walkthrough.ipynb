# Irisデータセットのロード
今回はscikit-learnからではなく、csvファイルからIristデータセットをロードします。
Iristデータセットは[こちら](http://living-with-data.jp/wp-content/uploads/2018/05/iris.csv)からダウンロードできます。
Pandasのread_csvという関数でファイルを読み込みます。

```python
df = pd.read_csv("iris.csv")
```

headとtailでデータフレームの最初の数行と最後の数行を表示できます。

```python
df.head()
```

```python
df.tail()
```

便利機能としてデータフレームの基本的な統計データを表示させる関数としてdescribeが用意されています。

```python
df.describe()
```

残念ながら上記のdescribeではカテゴリーデータについては無視されるので、例えばNameのユニークの値が知りたい場合は次のようにします。

```python
df.Name.unique()
```

Pandasには、簡単なプロット機能もあります。これはmatplotlibやseabornを使ってもよいのですが、Pandasでもできます。

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(df, alpha=0.2, figsize=(12, 12), diagonal='kde', color="red")
```

少し変わったプロットとしてandrews curveを描くことができます。

```python
from pandas.plotting import andrews_curves
andrews_curves(df, 'Name')
```
